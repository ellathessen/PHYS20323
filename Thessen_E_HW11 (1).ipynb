```python
class ChargedParticle:
    def __init__(self, x, y, z, vx, vy, vz, charge):
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
        self.charge = charge

    def move(self):
        self.x += self.vx
        self.y += self.vy
        self.z += self.vz

    def interact(self, other):
        if abs(self.x - other.x) < 0.25 and abs(self.y - other.y) < 0.25 and abs(self.z - other.z) < 0.25:
            if self.charge != other.charge:
                print("merge")
                self.vx = 0
                self.vy = 0
                self.vz = 0
                other.vx = 0
                other.vy = 0
                other.vz = 0
                self.charge = 0
                other.charge = 0
            else:
                print("repel")
                self.vx = -self.vx
                self.vy = -self.vy
                self.vz = -self.vz
                other.vx = -other.vx
                other.vy = -other.vy
                other.vz = -other.vz


p1 = ChargedParticle(-5, -5, -5, 1, 1, 1, 0.5)
p2 = ChargedParticle(5, 5, 5, -1, -1, -1, -0.5)

print("Initial positions and velocities:")
print(f"P1: ({p1.x}, {p1.y}, {p1.z}), Vel: ({p1.vx}, {p1.vy}, {p1.vz})")
print(f"P2: ({p2.x}, {p2.y}, {p2.z}), Vel: ({p2.vx}, {p2.vy}, {p2.vz})")

p1.move()
p2.move()
p1.interact(p2)

print("After moving:")
print(f"P1: ({p1.x}, {p1.y}, {p1.z}), Vel: ({p1.vx}, {p1.vy}, {p1.vz})")
print(f"P2: ({p2.x}, {p2.y}, {p2.z}), Vel: ({p2.vx}, {p2.vy}, {p2.vz})")
```

```python
 class ChargedParticle:
    def __init__(self, x, y, z, vx, vy, vz, charge):
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
        self.charge = charge

    def move(self, dt):
        self.x += self.vx * dt
        self.y += self.vy * dt
        self.z += self.vz * dt

    def interact(self, other):
        if abs(self.x - other.x) < 0.25 and abs(self.y - other.y) < 0.25 and abs(self.z - other.z) < 0.25:
            if self.charge != other.charge:
                print("merge")
                self.vx = self.vy = self.vz = 0
                other.vx = other.vy = other.vz = 0
                self.charge = other.charge = 0
            else:
                print("repel")
                self.vx = -self.vx
                self.vy = -self.vy
                self.vz = -self.vz
                other.vx = -other.vx
                other.vy = -other.vy
                other.vz = -other.vz

p1 = ChargedParticle(-5, -5, -5, 2, 2, 2, 0.5)
p2 = ChargedParticle(5, 5, 5, -1, -1, -1, 2.0)


dt = 0.05
steps = 300

for step in range(steps):
    
    print(f"Step {step+1}:")
    print(f"P1: ({p1.x}, {p1.y}, {p1.z}), Vel: ({p1.vx}, {p1.vy}, {p1.vz})")
    print(f"P2: ({p2.x}, {p2.y}, {p2.z}), Vel: ({p2.vx}, {p2.vy}, {p2.vz})")
    
    p1.move(dt)
    p2.move(dt)
    p1.interact(p2)
```
